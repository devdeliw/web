### Experiment 1 -- Law of Reflection

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {'Deval': pd.Series([5.5, 20.3, 40.5, 60, 80],
                      index=['5', '20', '40', '60', '80']),
        'Michaela': pd.Series([5.1, 20.1, 41, 60, 80],
                      index=['5', '20', '40', '60', '80']),
        'SuryaNeil': pd.Series([5.05, 20.5, 40, 60, 80],
                      index=['5', '20', '40', '60', '80']), 
        'Theoretical': pd.Series([5, 20, 40, 60, 80], 
                      index=['5', '20', '40', '60', '80'])}

df = pd.DataFrame(data)
df
```

###### Calculating Chi Squared

```python
d = df['Deval']
m = df['Michaela']
s = df['SuryaNeil']
t = df['Theoretical']
```

```python
fig, axis = plt.subplots(1, 1, figsize = (20,10))

axis.plot(t, d, 'r+', label = "Deval")
axis.plot(np.unique(t), np.poly1d(np.polyfit(t, d, 1))(np.unique(t)), label = "Deval")
axis.plot(t, m, 'g+', label = "Michaela")
axis.plot(np.unique(t), np.poly1d(np.polyfit(t, m, 1))(np.unique(t)), label = "Michaela")
axis.plot(t, s, 'b+', label = "SuryaNeil")
axis.plot(np.unique(t), np.poly1d(np.polyfit(t, s, 1))(np.unique(t)), label = "SuryaNeil")
axis.plot(t, t, 'k+', label = "Deval")
axis.plot(np.unique(t), np.poly1d(np.polyfit(t, t, 1))(np.unique(t)), label = "Theoretical")

y_error = 0.5
x_error = 0.5

plt.errorbar(t, d,
             yerr = y_error,
             xerr = x_error,
             fmt ='o', ecolor = 'r')
plt.errorbar(t, m,
             yerr = y_error,
             xerr = x_error,
             fmt ='o', ecolor = 'g')
plt.errorbar(t, s,
             yerr = y_error,
             xerr = x_error,
             fmt ='o', ecolor = 'b')

axis.legend()
axis.set_xlabel('incident angle (degrees)')
axis.set_ylabel('reflected angle (degrees)')
axis.set_title('incident vs. reflected angle -- experiment 1')
```

```python
residuals_squared_sum = 0

for i in range(len(d)): 
    residuals_squared_sum += (d.iloc[i] - t.iloc[i])**2 + (m.iloc[i] - t.iloc[i])**2 + (s.iloc[i] - t.iloc[i])**2

chi_square = residuals_squared_sum / 0.5**2
# the number of degrees of freedom is 14

reduced_chi_squared = chi_square/ 14

print(chi_square)
print(reduced_chi_squared)
```
